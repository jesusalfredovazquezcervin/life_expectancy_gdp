# Importing python modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

# Data Dictionary

1.- Country: The country of the record(Object)

2.- Year: Year of the record(Int64)

3.- Life expectancy at birth(years)(float64)

4.- GDP: Gross Domestic Product(float64)

# Loading data...

Loading the file "all_data.csv"

```python
led = pd.read_csv('all_data.csv')
print(led.info())
```

# Exploring data...

```python
print(led.head())
print(led.Country.unique())
```

# Changing column name...

```python
led.rename(columns={'Life expectancy at birth (years)':'Expectancy'}, inplace=True)
print(led.info())
```

# Has life expectancy increased over time in the six nations?

```python
#sns.scatterplot(rentals.sisqft, rentals.rent, hue = rentals.borough, palette='bright')
#plt.show()
fig, axs = plt.subplots(1, figsize=(12, 8))
sns.lineplot(x="Year", y="Expectancy", hue="Country", data=led, ax=axs)
sns.set_style("whitegrid")
plt.savefig("life_expectancy_increased_over_time.png")
plt.show()
plt.clf()
print("\nYes, the life expectancy has increased over the time for all six countries!")
```

# Has GDP increased over time in the six nations?

```python
fig, axs = plt.subplots(1, figsize=(13, 7))
sns.scatterplot(x = 'Year', y = 'GDP', data = led, hue='Country', ax = axs)
sns.set_style("whitegrid")
sns.color_palette('Dark2')
plt.show()
plt.clf()
print("\nThe GDP has increased over time, except CHILE & ZIMBABWE.")
```

# Is there a correlation between GDP and life expectancy of a country?

```python
fig, axs = plt.subplots(1, figsize=(13, 7))
sns.scatterplot(x=led.GDP, y=led.Expectancy, hue = led.Country, ax=axs, data = led)
plt.savefig('correlation_gdp_life_expectancy.png')
plt.show()
plt.clf()
print("\nYes, aparently there is a correlation for all the countries except for Zimbabwe and Chile.\nChile and Zimbabwe yes indeed had an increase in life expectancy but in her GDP had almost null increase.")
```

# What is the average life expectancy in these nations?

```python
fig, axs = plt.subplots(1, figsize=(13, 7))
sns.boxplot(x = "Country", y = "Expectancy", data = led, ax=axs)
plt.show()
plt.clf()
```

# What is the distribution of that life expectancy?

```python
#Chile' 'China' 'Germany' 'Mexico' 'United States of America' 'Zimbabwe'
sns.set_style("white")
sns.histplot(x='Expectancy',data=led)
plt.show()
plt.clf()

fig, axs = plt.subplots(2, 3, figsize=(15, 8))
#Row 1
axs[0,0].set_xlabel('Expectancy Chile')
axs[0,1].set_xlabel('Expectancy China')
axs[0,2].set_xlabel('Expectancy Germany')
sns.histplot(x='Expectancy',data=led[led.Country=='Chile'], ax=axs[0,0])
sns.histplot(x='Expectancy',data=led[led.Country=='China'], ax=axs[0,1])
sns.histplot(x='Expectancy',data=led[led.Country=='Germany'], ax=axs[0,2])


#Row 2
axs[1,0].set_xlabel('Expectancy Mexico')
axs[1,1].set_xlabel('Expectancy USA')
axs[1,2].set_xlabel('Expectancy Zimbabwe')
sns.histplot(x='Expectancy',data=led[led.Country=='Mexico'], ax=axs[1,0])
sns.histplot(x='Expectancy',data=led[led.Country=='United States of America'], ax=axs[1,1])
sns.histplot(x='Expectancy',data=led[led.Country=='Zimbabwe'], ax=axs[1,2])



plt.savefig('life_expectancy_distribution.png')
plt.show()
plt.clf()

```
